```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from a CSV file
def load_dataset(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except Exception as e:
        print(f"Error loading dataset: {str(e)}")
        return None

# Create bar chart for attack types with vertical labels, without showing counts
def create_attack_type_bar_chart(df):
    plt.figure(figsize=(14, 8))  # Adjust size to fit all labels
    sns.set(style="whitegrid")
    
    # Create the bar plot without adding counts
    ax = sns.countplot(x="AttackType", data=df, palette="Set2")
    
    # Rotate labels to be vertical
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    
    # Customize the plot
    plt.title("Attack Type Distribution")
    plt.xlabel("Attack Type")
    plt.ylabel("Count")
    plt.tight_layout()  # Ensures labels fit well
    plt.show()

# Create scatter plot for data length and attack type
def create_data_length_attack_type_scatter_plot(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="Data_Length", y="AttackType", data=df)
    plt.title("Data Length vs Attack Type")
    plt.xlabel("Data Length")
    plt.ylabel("Attack Type")
    plt.show()

# Create histogram for data length
def create_data_length_histogram(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df["Data_Length"], bins=10)
    plt.title("Data Length Distribution")
    plt.xlabel("Data Length")
    plt.ylabel("Count")
    plt.show()

# Main function
def main():
    file_path = "E:/Capstone_IDS/Modbus_Dataset/balanced_modbus_data.csv"  # replace with your CSV file path
    df = load_dataset(file_path)
    
    if df is not None:
        create_attack_type_bar_chart(df)
        create_data_length_attack_type_scatter_plot(df)
        create_data_length_histogram(df)

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each model (Accuracy, Precision, Recall, F1-Score)
models = ['KNN', 'SVM', 'AdaBoost', 'Random Forest', 'Stacking Classifier']

metrics = {
    'Accuracy': [99.54, 99.34, 98.83, 99.52, 99.78],
    'Precision': [0.77, 0.50, 0.10, 0.77, 0.90],
    'Recall': [0.76, 0.55, 0.14, 0.78, 0.89],
    'F1-Score': [0.76, 0.50, 0.12, 0.78, 0.89]
}

# Number of variables
num_vars = len(metrics)

# Prepare the radar chart angles (one per metric)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Repeat the first angle to close the circle
angles += angles[:1]

# Prepare the figure
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each model's metrics on the radar chart
for model, color in zip(models, ['steelblue', 'indianred', 'olivedrab', 'orange', 'purple']):
    values = [metrics['Accuracy'][models.index(model)],
              metrics['Precision'][models.index(model)],
              metrics['Recall'][models.index(model)],
              metrics['F1-Score'][models.index(model)]]
    
    # Repeat the first value to close the loop
    values += values[:1]
    
    ax.fill(angles, values, color=color, alpha=0.25)
    ax.plot(angles, values, color=color, linewidth=2, label=model)

# Labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(list(metrics.keys()), fontsize=12)

# Set range for each metric to normalize (as accuracy is in percentage)
ax.set_yticklabels([])
ax.set_ylim(0, 100)  # Accuracy is out of 100

# Add a title and legend
plt.title('Comparison of Models on Accuracy, Precision, Recall, and F1-Score', size=15, color='darkblue', y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Display the radar chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file (update the file path as needed)
file_path = "E:/Capstone_IDS/Modbus_Dataset/balanced_modbus_data.csv"  # Replace with the path to your CSV file
data = pd.read_csv(file_path)

# Convert the Timestamp column to datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'], errors='coerce')

# Drop rows with invalid timestamps, if any
data.dropna(subset=['Timestamp'], inplace=True)

# Add a new column to differentiate traffic type for grouping
data['Traffic Type'] = data['AttackType'].apply(lambda x: 'Benign' if x == 'benign' else 'Attack')

# Create a new column for hour and day to aggregate data
data['Hour'] = data['Timestamp'].dt.hour
data['Date'] = data['Timestamp'].dt.date

# Create a pivot table for heatmap
heatmap_data = data.pivot_table(index='Hour', columns='Date', values='Traffic Type', 
                                 aggfunc=lambda x: (x == 'Attack').sum(), fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)

# Adding titles and labels for clarity
plt.title('Heatmap of Attack Frequency by Hour')
plt.xlabel('Date')
plt.ylabel('Hour of Day')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Display plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from a CSV file
file_path = 'E:/Capstone_IDS/Modbus_Dataset/balanced_modbus_data.csv'  # Change this to the path of your CSV file
data = pd.read_csv(file_path)

# Convert the 'Timestamp' column to datetime format (MM.S)
# Assuming the timestamp is in the format 'MM.S' (e.g., '24.41.04' means 24 minutes and 41 seconds)
# We need to convert it to a format that pandas can understand
data['Timestamp'] = pd.to_timedelta(data['Timestamp'].str.replace('.', ':'))

# Set the 'Timestamp' column as the index for time series analysis
data.set_index('Timestamp', inplace=True)

# Resample the data to count the number of attacks per time interval (e.g., per minute)
attack_counts = data.resample('T').size()

# Plot the number of attacks over time
plt.figure(figsize=(12, 6))
plt.plot(attack_counts.index, attack_counts.values, label='Total Attacks', color='blue')
plt.title('Number of Attacks Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# Count the frequency of different attack types over time
attack_type_counts = data.groupby([pd.Grouper(freq='T'), 'AttackType']).size().unstack(fill_value=0)

# Plot the frequency of different attack types over time
plt.figure(figsize=(12, 6))
attack_type_counts.plot(kind='line', marker='o')
plt.title('Frequency of Different Attack Types Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.grid()
plt.legend(title='Attack Type')
plt.tight_layout()
plt.show()
```
