```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE

# Load the data
file_path = "E:/Capstone_IDS/Modbus_Dataset/balanced_modbus_data.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print("Initial DataFrame:")
print(df.head())

# Preprocess the data
# Fill missing values
df.fillna({
    'Source_IP': 'unknown',
    'Destination_IP': 'unknown',
    'Source_Port': 0,
    'Destination_Port': 0,
    'Protocol': 'unknown',
    'Data_Length': 0,
    'TargetIP': 'unknown',
    'TransactionID': 0
}, inplace=True)

# Convert timestamp to a numeric feature (e.g., seconds from start)
df['Timestamp'] = df['Timestamp'].str.split(':', expand=True).astype(float).apply(lambda x: x[0]*60 + x[1], axis=1)

# Encode categorical features
label_encoders = {}
for column in ['Source_IP', 'Destination_IP', 'Protocol', 'AttackType', 'TargetIP']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Separate the data
# Excluded values for AttackType
excluded_values = ['benign', 'Brute force or specific coil. Address: 1', 'Brute force or specific - Complete']
# Transform excluded_values using the same encoder
excluded_values_encoded = label_encoders['AttackType'].transform(excluded_values)

# Separate data into excluded and SMOTE-applicable datasets
df_excluded = df[df['AttackType'].isin(excluded_values_encoded)]
df_smote = df[~df['AttackType'].isin(excluded_values_encoded)]

# Separate features and target for SMOTE
X_smote = df_smote.drop('AttackType', axis=1)
y_smote = df_smote['AttackType']

# Standardize the feature set (excluding the target column)
scaler = StandardScaler()
X_smote_scaled = scaler.fit_transform(X_smote)

# Apply SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_smote_scaled, y_smote)

# Convert resampled data back to DataFrame
df_resampled = pd.DataFrame(X_resampled, columns=X_smote.columns)
df_resampled['AttackType'] = y_resampled

# Combine the excluded data with the resampled data
df_final = pd.concat([df_excluded, df_resampled], ignore_index=True)

# Print the first few rows of the final dataframe
print("Final DataFrame:")
print(df_final.head())

# Optionally, save the final DataFrame
df_final.to_csv("E:/Capstone_IDS/Modbus_Dataset/resampled_modbus_data.csv", index=False)
```
